## Get the Statistics of Dataframe, Checking Missing Value

```python
def data_descr(data, data_name=''):
    print(f'The dataset is: {data_name}', end='\n\n')
    display(data.info())
    
    print(f'Statistical information', end='\n\n')
    display(data.describe(include='all').T)
    
    duplicates = data.duplicated().sum()
    if duplicates > 0:
        print('The number of duplicates is', data.duplicated().sum(), '.')
    else:
        print('There are no duplicates in the data')
        
    print('The number of missing values per column in % ')
    report = data.isna().sum() / data.shape[0] * 100
    report = report.to_frame()
    report.rename(columns={0:'missing values in % of total'}, inplace =True)
    display(report)
    
data_descr(df, data_name="Stroke Dataset")
```

```python
import keras
from keras import callbacks
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import BatchNormalization
from keras.optimizers import Adam

# Early stopping
early_stopping = callbacks.EarlyStopping(
    min_delta=0.001,
    patience=20,
    restore_best_weights=True,
)

# Initialing the NN with creating a sequential model
model = Sequential()

# adding the first input layer and the first hidden layer
# model.add(Dense(15, kernel_initializer = 'normal', activation = 'relu', input_shape = (6, )))
model.add(Dense(1024, kernel_initializer='normal', activation = 'relu', input_shape = (6, )))

# adding batch normalization and dropout layer
model.add(Dropout(rate = 0.1))
# model.add(Dropout(p=0.1))
model.add(BatchNormalization())


model.add(Dense(512, kernel_initializer='normal', activation = 'relu'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(256, kernel_initializer='normal', activation = 'relu'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(128, kernel_initializer='normal', activation = 'relu'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(64, kernel_initializer='normal', activation = 'relu'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(32, kernel_initializer='normal', activation = 'relu'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(16, kernel_initializer='normal', activation = 'relu'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

# adding the third hidden layer
# model.add(Dense(7, kernel_initializer = 'normal', activation = 'relu'))
model.add(Dense(8, kernel_initializer='normal', activation = 'relu'))

# adding batch normalization and dropout layer
model.add(Dropout(rate = 0.1)) 
# model.add(Dropout(p=0.1))
model.add(BatchNormalization())

# adding the output layer
# model.add(Dense(2, kernel_initializer = 'normal', activation = 'relu')
model.add(Dense(2, kernel_initializer = 'normal', activation = 'relu'))

# Compiling the ANN
# model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
opt = Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                      callbacks=[early_stopping],  
                      validation_split = 0.2, 
                      validation_data = (X_test, y_test), 
                      epochs = 10)
```
