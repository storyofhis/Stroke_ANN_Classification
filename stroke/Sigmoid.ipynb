```python
import pandas as pd 

df = pd.read_csv("health_data.csv")
df.drop(['Diabetes', 'Hypertension'], axis=1, inplace=True)
df
```

## Get the Statistics of Dataframe, Checking Missing Value

```python
def data_descr(data, data_name=''):
    print(f'The dataset is: {data_name}', end='\n\n')
    display(data.info())
    
    print(f'Statistical information', end='\n\n')
    display(data.describe(include='all').T)
    
    duplicates = data.duplicated().sum()
    if duplicates > 0:
        print('The number of duplicates is', data.duplicated().sum(), '.')
    else:
        print('There are no duplicates in the data')
        
    print('The number of missing values per column in % ')
    report = data.isna().sum() / data.shape[0] * 100
    report = report.to_frame()
    report.rename(columns={0:'missing values in % of total'}, inplace =True)
    display(report)
    
data_descr(df, data_name="Stroke Dataset")
```

```python
import matplotlib.pyplot as plt 

values = df.Stroke.value_counts()
labels = ['Not Stroke', 'Stroke']

fig, ax = plt.subplots(figsize = (4, 3), dpi = 100)
explode = (0, 0.09)

patches, texts, autotexts = ax.pie(values, labels = labels, autopct = '%1.2f%%', shadow = True, 
                                   startangle = 90, explode = explode)
plt.setp(texts, color = 'grey')
plt.setp(autotexts, size = 8, color = 'white')
autotexts[1].set_color('black')
plt.show()
```

```python
from imblearn.over_sampling import SMOTENC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, KFold, cross_validate

# spliting model
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
X = df.loc[:, ['Age', 'HighChol', 'HeartDiseaseorAttack', 'GenHlth', 'PhysHlth', 'DiffWalk']]
y = to_categorical(df.Stroke)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state = 1248)

# Data balancing
oversample = SMOTENC([1,2,3,4,5], random_state=1248)
X,y = oversample.fit_resample(X_train, y_train)

# df = pd.concat([X,y], axis=1)
# df
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
# print(df.shape)
```

```python
# Scaling data with standard scaler
# the Standard score of a sample x is calculated as : 
# z = (x - u)/s 
# u : Mean of training samples or zero if with_mean = False.
# s = standard deviation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train) # fit to data, then transform it -> return a transformed version of X.
X_test = sc.transform(X_test)       # Perform standardization by centering and scalling
```

```python
import keras
from keras import callbacks
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import BatchNormalization
from keras.optimizers import Adam

# Early stopping
early_stopping = callbacks.EarlyStopping(
    min_delta=0.001,
    patience=20,
    restore_best_weights=True,
)

# Initialing the NN with creating a sequential model
model = Sequential()

# adding the first input layer and the first hidden layer
# model.add(Dense(15, kernel_initializer = 'normal', activation = 'relu', input_shape = (6, )))
model.add(Dense(1024, kernel_initializer='normal', activation = 'sigmoid', input_shape = (6, )))

# adding batch normalization and dropout layer
model.add(Dropout(rate = 0.1))
# model.add(Dropout(p=0.1))
model.add(BatchNormalization())


model.add(Dense(512, kernel_initializer='normal', activation = 'sigmoid'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(256, kernel_initializer='normal', activation = 'sigmoid'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(128, kernel_initializer='normal', activation = 'sigmoid'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(64, kernel_initializer='normal', activation = 'sigmoid'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(32, kernel_initializer='normal', activation = 'sigmoid'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

model.add(Dense(16, kernel_initializer='normal', activation = 'sigmoid'))
# model.add(Dropout(rate = 0.1)) 
model.add(BatchNormalization())

# adding the third hidden layer
# model.add(Dense(7, kernel_initializer = 'normal', activation = 'relu'))
model.add(Dense(8, kernel_initializer='normal', activation = 'sigmoid'))

# adding batch normalization and dropout layer
model.add(Dropout(rate = 0.1)) 
# model.add(Dropout(p=0.1))
model.add(BatchNormalization())

# adding the output layer
# model.add(Dense(2, kernel_initializer = 'normal', activation = 'relu')
model.add(Dense(2, kernel_initializer = 'normal', activation = 'sigmoid'))

# Compiling the ANN
# model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
opt = Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                      callbacks=[early_stopping],  
                      validation_split = 0.2, 
                      validation_data = (X_test, y_test), 
                      epochs = 10)
```

```python
history_df = pd.DataFrame(history.history)

plt.figure(figsize = (12, 6))

train_loss = history_df.loc[:, ['loss']]
val_loss = history_df.loc[:, ['val_loss']]
# epoch = range(1, 101)
plt.plot( train_loss, "#BDE2E2",  label = 'Training Loss')
plt.plot( val_loss, "#C2C4E2", label = 'Validation Loss')
plt.title('Training and Validation Loss\n')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
history_df = pd.DataFrame(history.history)
plt.figure(figsize = (12, 6))

train_accuracy = history_df.loc[:, ['accuracy']]
val_accuracy = history_df.loc[:, ['val_accuracy']]
# epoch = range(1, 101)
plt.plot( train_accuracy, "#BDE2E2",  label = 'Training accuracy')
plt.plot( val_accuracy, "#C2C4E2", label = 'Validation accuracy')
plt.title('Training and Validation Accuracy\n')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
eva = model.evaluate(X_train, y_train)[1]

print(f'Accuracy of model is {eva}')
```

```python
acc = model.evaluate(X_test, y_test)[1]

print(f'Accuracy of model is {acc}')
```
